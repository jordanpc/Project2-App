```python
drinking_df = pd.read_csv("drinking_age.csv")
```

```python
drinking_df = drinking_df.drop(columns=['Unnamed: 3','Unnamed: 4','Unnamed: 5','Unnamed: 6','Unnamed: 7'])
drinking_df.head()
```

```python
new_combined_df = combined_df.merge(drinking_df, how='outer', on="Country")
```

```python
new_combined_df = new_combined_df[:190]
```

```python
new_combined_df.head()
```

```python
death_df = pd.read_csv("DeathRate.csv")
```

```python
death_df = death_df.rename(columns={"Country Name":"Country"})
```

```python
final_df = new_combined_df.merge(death_df, how='outer', on="Country")
```

```python
final_df = final_df.dropna(how="all", subset=["Death Rate Per 100,000"]).reset_index(drop=True)
```

```python
final_df = final_df[:180]
```

```python
final_df = final_df.rename(columns={'Recorded Consumption':'Recorded_Consumption',
                                    'Unrecorded Consumption':'Unrecorded_Consumption',
                                    'Beer (%)':'Beer_Percent',
                                    'Wine (%)':'Wine_Percent',
                                    'Spirits (%)':'Spirits_Percent',
                                    'Other (%)':'Other_Percent',
                                    'GDP (US$MM)':'GDP',
                                    'On Premise Purchase Age':'Purchase_Age',
                                    'Country Code':'iso_a3',
                                    'Death Rate Per 100,000':'Death_Rate'})
final_df = final_df[["Country","iso_a3","Total","Recorded_Consumption","Unrecorded_Consumption","Beer_Percent","Wine_Percent","Spirits_Percent","Other_Percent","GDP","Purchase_Age","Death_Rate"]]
final_df
```

```python
final_df.dtypes
```

```python
engine = create_engine("sqlite:///alcoholDB.sqlite")
```

```python
Base = declarative_base()

class AlcoholClass(Base):
    __tablename__ = 'alcohol'
    id = Column(Integer, primary_key=True)
    Country = Column(String(100))
    Total = Column(Float)
    Recorded_Consumption = Column(Float)
    Unrecorded_Consumption = Column(Float)
    Beer_Percent = Column(Float)
    Wine_Percent = Column(Float)
    Spirits_Percent = Column(Float)
    Other_Percent = Column(Float)
    GDP = Column(Float)
    Purchase_Age = Column(Float)
    Country_Code = Column(String(50))
    Death_Rate = Column(Float)

Base.metadata.create_all(engine)
```

```python
final_df.to_sql('alcohol', con = engine, if_exists='append', index=False)
```
